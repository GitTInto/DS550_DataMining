## 1. Create an array of size [10,50] and use the random function to fill it with integers between 1 and 500 (inclusive). Load it to a DataFrame.

```python
# Load library
import numpy as np
import pandas as pd
```

```python
# Generating 500 random numbers between 1 and 500
# Shape them into 10*50
# load that into a dataframe.


#np.random.randint(1, 501)
#array = np.random.randint(0, 16,15)
array = np.random.randint(1, 501, 500)
#x = array.reshape(3,5)
x = array.reshape(10,50)
df = pd.DataFrame.from_records(x)
#df = pd.DataFrame(array)
df
```

## 2. Calculate the sum of each row; the sum of each column; and the sum of all entries.

```python
# Sum of each Row
df.sum(axis = 1)
```

```python
# Sum of each Column
df.sum(axis = 0)
```

```python
# Sum of all entries

#def sumof(x):
#    return x.sum()

k = df.apply(lambda x: x.sum())
k.sum(axis = 0)
```

```python
# Another way of doing Sum of all entries
df1 = df.sum(axis = 1)
df1.sum(axis = 0)
```

```python

# Load library
import pandas as pd
import matplotlib.pyplot as plt
```

```python
url = './DS_Vs_ML_Search.csv'

# If a header row does not exist, we set header=None.
# header = None

# Load dataset
dataframe = pd.read_csv(url)

# View first two rows
dataframe.head(2)
```

```python
# To view the full data frame
#dataframe
```

```python
# converting these into arrays
#All of plotting functions expect np.array or np.ma.masked_array
# as input. Classes that are 'array-like' such as pandas data objects
# and np.matrix may or may not work as intended.
# It is best to convert these to np.array objects prior to plotting.

data_asarray = dataframe.values
#data_asarray
week = dataframe['Week'].values
Data_science = dataframe['Data science'].values
Machine_learning = dataframe['Machine learning'].values
```

```python
plt.plot(week, Data_science, color='blue', marker='o', label='Data science')
plt.plot(week, Machine_learning, color='red', marker='x' ,label='Machine learning')
plt.title('Google Searches', fontsize=14)
plt.xlabel('Week', fontsize=14)
plt.ylabel('Data science Vs Machine learning', fontsize=14)
plt.legend()
```
